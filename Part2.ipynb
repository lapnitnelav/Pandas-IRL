```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as p
```

```python
dfb=pd.read_csv('2013---2015-dfb-ambulance.csv')
dfb['DTStart']=dfb[['Date','TOC']].apply(lambda x: ' '.join(x), axis=1)
```

We start off with the classic MPL magic, the imports and reading the csv file.

We also compute a full DateTime by joining Date and TOC :

```python
dfb.head()
```

```python
dfb['DTStart']=pd.to_datetime(dfb['DTStart'],format="%d/%m/%Y %H:%M:%S")
```

Once converted to DateTime, we extract the week and day of week.

```python
dfb['Week']=dfb['DTStart'].dt.week
dfb['Day']=dfb['DTStart'].dt.weekday_name
dfb
```

```python
week_day_view=dfb['TOC'].groupby([dfb['Day'],dfb['Week']]).count()
week_day_view.head(5)
```

```python
p.style.use('fivethirtyeight')
```

```python
days_by_week=week_day_view.unstack('Day')
days_by_week.head(5)
```

Nothing new here, except the obvious issue in the ordering of columns, due to unstack auto sorting.

While it doesn't create any problem for crunching data, it's a bit less easy on the eye, especially for graphs

Luckily, fixing that is fairly easy :

```python
cols=days_by_week.columns
reorder=[1,5,6,4,0,2,3]
days_by_week.columns=[cols[index] for index in reorder]
days_by_week.head(5)
```

First extract the current columns, then generate the correct index order (0-based) list.
This allows us to use a list comprehension which is going to go through the reorder list and use the index to pick from the columns in the right order.


This could be rewritten as :

```python
ordered_cols=[]
for index in reorder:
    ordered_cols.append(cols[index])
ordered_cols
#To actually assign the change :
#days_by_week.columns=ordered_cols
```

Plotting that is quite straighforward, I'm only adding a few parameters for MLP, namely using a background grid, defining a custom range for the y axis (so the legend doesn't block the view) and setting 20% transparency :

```python
days_by_week.plot.area(figsize=(18,12),grid=True,yticks=range(0,7000,500),alpha=0.8)
```

But this is hardly practical to look at variations on the volume per day, so we are going to transform the data again to get what we need, which is a % rather than absolute values.


First off, summing up for each week:

```python
days_by_week['Total']=days_by_week.sum(axis=1)
days_by_week.head(5)
```

Then we do a simple math operation, Day of Week / Total, for each day in the week by first spawning a new DataFrame and iterating over each day (reusing existing columns except 'Total') and reordering.

```python
pct_days_week=pd.DataFrame()
for i in cols :
    pct_days_week[i]=days_by_week[i]/days_by_week['Total']
pct_days_week.columns=[cols[index] for index in reorder]
pct_days_week.head(5)
```

```python
pct_days_week.plot.area(figsize=(18,12),yticks=[float((x+1))/20 for x in range(20)],grid=True,alpha=0.8,colormap='winter')
p.legend(ncol=4)
```

New options regarding plots :

* Colormap to get a broader range of colours than the 5 basic ones and avoid overlap. See this [chart](http://scipy.github.io/old-wiki/pages/Cookbook/Matplotlib/Show_colormaps) to choose / experiment

* changing number of columns in the legend for a horizontal display

* also, generating a list from 0 to 1 by 0.05 increment for the yaxis.

```python
import numpy as np
pct_days_week.mean().plot(figsize=(12,8),yticks=np.linspace(0.135,0.155,9))
p.axhline(y=0.1428)
```

Here we replace the cumbersome list comprehension by Numpy's [linspace](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html#numpy.linspace), an alternative which can be easier on the brain.
We also use MLP's axhline (vertical counterpart is axvline) to add the mean as a static value.

---
It appears that Friday and Sunday are the busiest while midweek is probably more 'relaxed'.


Now let's look into the variance for each day. We could easily use a bar graph (combined with mean and std) but another way to examine that would be to use an histogram:


```python
days_by_week[cols].hist(figsize=(15,14),bins=np.linspace(450,700,11))
```

Ok, proper order went out the door again and I'm not sure there's an easy way to fix it so let's leave that aside.

On the creating the histogram, nothing too complicated here: select the columns (so mostly to leave out Total), and provide the bins we want to use. Bins are intervals (=range), ie from 600 to 625. And we get to see frequency for each bin, which means how often each range pops up.

For example Sunday seems fairly consistent about 550 to 575 incidents per day whereas Thursday is a bit all over the place. Pay attention to the Y axis which is different for each graph.

---

For the sake of completeness, we could ensure everything shows up in the same graph but I find it fairly confusing to read, which defeats the point of a visualisation.

```python
days_by_week[cols].plot.hist(figsize=(15,14),bins=np.linspace(450,700,11),alpha=0.8,colormap='jet')
```

Good luck making sense of that, even when playing with the various colour schemes.

Anyway, main change here is that we call plot.hist() instead of hist()

Let's move on and look at the activity per hour of the day.

```python
dfb['Hour']=dfb['DTStart'].dt.hour
hour_view=dfb['TOC'].groupby([dfb['Hour'],]).count()
hour_view.plot(figsize=(15,12),xticks=[float(x) for x in range(24)],yticks=np.linspace(4000,12000,17))
```

Nothing new here, but the graph itself is probably the most revealing so far.

Let's see if there's any variance per quarter :

```python
dfb['Quarter']=dfb['DTStart'].dt.quarter
hourly_per_quarter=dfb['TOC'].groupby([dfb['Hour'],dfb['Quarter']]).count().unstack(level=-1)
hourly_per_quarter
```

```python
hourly_per_quarter.plot(figsize=(15,12),xticks=[float(x) for x in range(24)])
```
